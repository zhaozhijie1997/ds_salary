### Salary Extraction

```python
df = df[df['Salary Estimate'] != '-1']
df.shape
```

```python
df['hourly'] = 0
df.loc[df['Salary Estimate'].str.contains('Per Hour'),'hourly'] = 1
df['employer provided'] = 0
df.loc[df['Salary Estimate'].str.contains('employer provided'),'employer provided'] = 1
```

```python
df.head(10)
```

```python
salary = df['Salary Estimate'].apply(lambda x:x.split('(')[0])
minus_kd = salary.apply(lambda x: x.replace('K','').replace('$',''))
min_hr = minus_kd.apply(lambda x: x.lower().replace('per hour','').replace('employer provided salary:',''))
```

```python
df['min_salary'] = min_hr.apply(lambda x:int(x.split('-')[0]))
df['max_salary'] = min_hr.apply(lambda x:int(x.split('-')[1]))
df['avg_salary'] = (df['min_salary'] + df['max_salary'] )/2
```

### Company Name

```python
df['Company Name'] = df['Company Name'].apply(lambda x:x.split('\n')[0])
```

```python
df[df['Company Name'].str.contains('\d')]
```

### State Field

```python
df['State'] = df['Location'].apply(lambda x:x.split(', ')[1])
```

```python
df['Location'].isnull().sum()
```

```python
df['Location'].value_counts()
```

### Create Columns to show if a job is at its headquarter

```python
df['AT_HQ'] = 0
```

```python
df.loc[df['Headquarters'] == df['Location'],'AT_HQ'] = 1
```

```python
df.head()
```

### Age of the company

```python
df['Age'] = df['Founded'].apply(lambda x:2020 - int(x) if x!= -1 else -1)
```

### Job Description

Relevant terms for data scientist roles
- Python
- R studio
- Spark
- AWS
- Excel

```python
df.loc[df['Job Description'].str.lower().str.contains('python'),'Python'] = 1
df.loc[df['Python'].isnull(),'Python'] = 0
```

```python
df['R Studio'] = df['Job Description'].apply(lambda x: 1 if 'r studio' in x.lower() or 'r-studio' in x.lower() else 0 )
```

```python
df['Spark'] = df['Job Description'].apply(lambda x: 1 if 'spark' in x.lower() else 0 )
```

```python
df['AWS'] = df['Job Description'].apply(lambda x: 1 if 'aws' in x.lower() else 0 )
```

```python
df['Excel'].value_counts()
```

```python
df['Excel'] = df['Job Description'].apply(lambda x: 1 if 'excel' in x.lower() else 0 )
```

```python
df.drop('Unnamed: 0',axis=1,inplace =True)
```

```python
df.to_csv('data_cleaned.csv',index=False)
```
